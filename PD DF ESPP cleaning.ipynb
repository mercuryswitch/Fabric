```python
# Code generated by Data Wrangler for PySpark DataFrame

def clean_data(df):
    # Filter rows based on column: 'Reporting_type'
    df = df.filter(~df['Reporting_type'].eqNullSafe("Row not used"))
    # Filter rows based on column: 'Row_number'
    df = df.filter(df['Row_number'].isNotNull())
    return df

df_clean = clean_data(df)
display(df_clean)
```
