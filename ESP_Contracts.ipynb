```python
# Load Contract notice file into Dataframe, clean and export for Sharepoint list


df = spark.read.format("csv").option("header","true").load("Files/Contract Search Export_Summary Export_data_15.07.24.csv")
# df now is a Spark DataFrame containing CSV data from "Files/Contract Search Export_Summary Export_data_15.07.24.csv".
display(df)
```

```python
# Code generated by Data Wrangler for pandas sample

def clean_data(pandas_df):
    # Select columns: '02. Agency Name (Summary Export)', '03. Contract Notice ID (Summary Export)', '06. Publish Date (Summary Export)'
    pandas_df = pandas_df.loc[:, ['02. Agency Name (Summary Export)', '03. Contract Notice ID (Summary Export)', '06. Publish Date (Summary Export)']]
    return pandas_df

# Loaded variable 'df' from kernel state
pandas_df = df.limit(5000).toPandas()

pandas_df_clean = clean_data(pandas_df.copy())
display(pandas_df_clean)


```

```python
# Code generated by Data Wrangler for pandas DataFrame

def clean_data(pandas_df_clean):
    # Rename column '03. Contract Notice ID (Summary Export)' to 'Contract Notice ID'
    pandas_df_clean = pandas_df_clean.rename(columns={'03. Contract Notice ID (Summary Export)': 'Contract Notice ID'})
    # Rename column '02. Agency Name (Summary Export)' to 'Department'
    pandas_df_clean = pandas_df_clean.rename(columns={'02. Agency Name (Summary Export)': 'Department'})
    # Rename column '06. Publish Date (Summary Export)' to 'Publish Date'
    pandas_df_clean = pandas_df_clean.rename(columns={'06. Publish Date (Summary Export)': 'Publish Date'})
    return pandas_df_clean

pandas_df_clean_1 = clean_data(pandas_df_clean.copy())
pandas_df_clean_1.head()
```

```python
import pandas as pd


pandas_df_clean_1.to_excel("/lakehouse/default/Files/ESP_Contracts.xlsx")
```
